```python
data_array = df.values # converting data frame to array
data_array
```

```python
# Defining X & Y variables
X = data_array[:,0:8] # Independent variables
Y = data_array[:,8] # Dependent variable
```

```python
# Splitting X, Y train & test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
# Standardizing X train & test data
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_train = SS.fit_transform(X_train)
X_test = SS.transform(X_test)
```

```python
X_train
```

```python
# knn ml model building
# from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, p=2) # 5 neighbours (ask 5 people make decision) & p= binary classification
knn.fit(X_train, Y_train)
```

```python
# predicting the test set results
Y_Predict = knn.predict(X_test)
Y_Predict
```

```python
# Creating the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_Predict)
cm
# [114+16+22+40 = 192 = (TotRec:768-train:75%(576))]
```

```python
# Model Evaluation (Accuracy)
accuracy = accuracy_score(Y_test, Y_Predict)
print("Accuracy:", accuracy)

# Key Targets:
# ≈ 50%	    No better than random guessing (binary classification)
# 60–70%	Weak / baseline model
# 70–85%	Reasonable / decent
# 85–95%	Very good
# > 95%	    Excellent (check for overfitting or data leakage)
# Note: High accuracy can be misleading if classes are imbalanced.
```

```python
# Classification Report
print(classification_report(Y_test, Y_Predict))

# Key Targets
# Precision > 0.7 (good)
# Recall > 0.7 (good)
# F1-score > 0.75 (strong)
```
